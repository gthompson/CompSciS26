# Constructing a Bode plot (Transfer function plot) from poles and zeros


```python
# These are poles and zeros and gain K, stated in Chaparral M25 manual
poles = [-1190, -0.157+3e-6j, -0.157-3e-6j, -0.157 ]
zeros = [0, 0, 0, -4080]
K = 0.2917
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Transfer function
system = signal.ZerosPolesGain(zeros, poles, K)

# Frequency range (rad/s)
w = np.logspace(-3, 4, 5000)
w, mag, phase = signal.bode(system, w=w)

# Convert rad/s to Hz
f = w / (2 * np.pi)

# Plot
fig, ax = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

ax[0].semilogx(f, mag)
ax[0].set_ylabel("Magnitude (dB)")
ax[0].grid(True, which="both")
ax[0].set_xlim(0.005, 400)
ax[0].set_ylim(-40, 0)
ax[0].set_title("Bode Plot of Chaparral M25")

ax[1].semilogx(f, phase)
ax[1].set_ylabel("Phase (degrees)")
ax[1].set_xlabel("Frequency (Hz)")
ax[1].grid(True, which="both")
ax[1].set_xlim(0.005, 400)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Your Chaparral M25 PZs (as given) ---
poles = np.array([-1190, -0.157 + 3e-6j, -0.157], dtype=complex)
zeros = np.array([0, 0, 0, -4080], dtype=complex)

def plot_pz_s_plane(poles, zeros, title="Pole-Zero Plot (s-plane)"):
    fig, ax = plt.subplots(figsize=(7, 6))

    # Plot zeros (o) and poles (x)
    if len(zeros):
        ax.scatter(np.real(zeros), np.imag(zeros), marker='o', s=80, facecolors='none', edgecolors='k', label='Zeros')
    if len(poles):
        ax.scatter(np.real(poles), np.imag(poles), marker='x', s=80, color='k', label='Poles')

    # Axes
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.set_xlabel("Real(s) [rad/s]")
    ax.set_ylabel("Imag(s) [rad/s]")
    ax.set_title(title)
    ax.grid(True, which="both")
    ax.legend()

    # Make plot roughly square in scale
    ax.set_aspect('equal', adjustable='datalim')
    plt.tight_layout()
    plt.show()

def s_to_z(p, fs, method="matched_z"):
    """
    Map s-plane poles/zeros to z-plane so the unit circle makes sense.

    matched_z (default): z = exp(s*dt)  (a.k.a. impulse-invariant / matched pole-zero mapping)
    bilinear:            z = (1 + s*dt/2) / (1 - s*dt/2)
    """
    dt = 1.0 / fs
    p = np.asarray(p, dtype=complex)

    if method == "matched_z":
        return np.exp(p * dt)
    elif method == "bilinear":
        return (1 + p*dt/2) / (1 - p*dt/2)
    else:
        raise ValueError("method must be 'matched_z' or 'bilinear'")

def plot_pz_z_plane(poles, zeros, fs, method="matched_z", title=None):
    zp = s_to_z(poles, fs, method=method)
    zz = s_to_z(zeros, fs, method=method)
    print(f"Mapped poles (z-plane): {zp}")
    print(f"Mapped zeros (z-plane): {zz}")

    fig, ax = plt.subplots(figsize=(7, 6))

    # Unit circle
    th = np.linspace(0, 2*np.pi, 600)
    ax.plot(np.cos(th), np.sin(th), linewidth=1, label="Unit circle")

    # Plot zeros/poles
    if len(zz):
        ax.scatter(np.real(zz), np.imag(zz), marker='o', s=80, facecolors='none', edgecolors='k', label='Zeros (mapped)')
    if len(zp):
        ax.scatter(np.real(zp), np.imag(zp), marker='x', s=80, color='k', label='Poles (mapped)')

    # Axes
    ax.axhline(0, linewidth=1)
    ax.axvline(0, linewidth=1)
    ax.set_xlabel("Real(z)")
    ax.set_ylabel("Imag(z)")

    if title is None:
        title = f"Pole-Zero Plot (z-plane), fs={fs} Hz, mapping={method}"
    ax.set_title(title)

    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-1.2, 1.2)
    ax.grid(True, which="both")
    ax.legend()
    plt.tight_layout()
    plt.show()

# --- 1) s-plane plot (correct native domain for your PZs) ---
plot_pz_s_plane(poles, zeros)

# --- 2) OPTIONAL: z-plane plot with unit circle (choose a sampling rate) ---
fs = 1000.0  # <-- set this to your actual sampling rate (Hz)
plot_pz_z_plane(poles, zeros, fs=fs, method="matched_z")
# plot_pz_z_plane(poles, zeros, fs=fs, method="bilinear")  # alternative
```
