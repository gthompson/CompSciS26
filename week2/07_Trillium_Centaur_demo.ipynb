**Goal (in class):** Set up a Trillium Compact PH + Nanometrics Centaur station.

---

## Checklist

- Trillium Compact PH broadband seismometer
- Centaur digitizer
- Centaur power cable
- GPS timing cable
- Trillium sensor cable(s) connected (from Trillium to Centaur, on Sensor A port)
- infrasound sensor(s)
- infrasound sensor cable (from infrasound to Centaur, on Sensor B port)
- SD card
- Ethernet cable
- laptop with ethernet port, or with a USB-ethernet adapter
- For the field:
  + plastic enclosure
  + solar panel
  + solar panel mount
  + battery (lead-acid, or LiFePO4)
  + solar charge controller ("regulator")
